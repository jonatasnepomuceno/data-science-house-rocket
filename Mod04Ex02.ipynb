## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
# 1) Seu código aqui

acertos_mat = df_mat['Qt_acertos'].sum() / (100 * 24)
print('Porcentagem acertos Mat:', acertos_mat)

acertos_por = df_por['Qt_acertos'].sum() / (100 * 18)
print('Porcentagem acertos Port:', acertos_por)

acertos_geo = df_geo['Qt_acertos'].sum() / (100 * 8)
print('Porcentagem acertos Geo:', acertos_geo)

acertos_ing = df_ing['Qt_acertos'].sum() / (100 * 8)
print('Porcentagem acertos Ing:', acertos_ing)

acertos_his = df_his['Qt_acertos'].sum() / (100 * 8)
print('Porcentagem acertos Hist:', acertos_his)

acertos_fis = df_fis['Qt_acertos'].sum() / (100 * 12)
print('Porcentagem acertos Fis:', acertos_fis)

acertos_qui = df_qui['Qt_acertos'].sum() / (100 * 12)
print('Porcentagem acertos Qui:', acertos_qui)


````

````python
# 2) Seu código aqui
##(operações entre *DataFrames) Calcule o total de acertos de cada aluno.
##  somar os dataframes
df_geral = df_mat + df_por + df_geo + df_ing + df_his + df_fis + df_qui
df_geral
````

````python
# 3) Seu código aqui
#Calcule o porcentual geral de cada aluno.
  
df_geral['% Geral'] = df_geral / 90 #numero de questões
df_geral
````

````python
# 4) Seu código aqui
##Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?
df_geral['Fase 2'] = 0
df_geral.loc[df_geral['Qt_acertos'] >= 45, 'Fase 2'] = 1
df_geral['Fase 2'].sum()
````
